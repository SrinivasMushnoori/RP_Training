```python
import time, math, multiprocessing
import matplotlib.pyplot as plt
import plotly.plotly as py
def f(n):
    
    for k in range(n):

        x = math.exp(50)
        y = math.exp(25)
        z = math.exp(100)



if __name__ == '__main__':
     
    t = 0

    for count in range(4):
        jobs = []
        start = time.time()
        for i in range(32):
            p = multiprocessing.Process(target=f, args=(67500,))
            jobs.append(p)
            p.start()
            p.join()

        end = time.time()
        
        t = t + (end - start)
    t32=t/4


    t = 0

    for count in range(4):
        jobs = []
        start = time.time()
        for i in range(16):
            p = multiprocessing.Process(target=f, args=(125000,))
            jobs.append(p)
            p.start()
            p.join()

        end = time.time()
        
        t = t + (end - start)
    t16=t/4

    t = 0

    for count in range(4):
        jobs = []
        start = time.time()
        for i in range(8):
            p = multiprocessing.Process(target=f, args=(250000,))
            jobs.append(p)
            p.start()
            p.join()

        end = time.time()
        
        t = t + (end - start)
    t8=t/4

    t = 0

    for count in range(4):
        jobs = []
        start = time.time()
        for i in range(4):
            p = multiprocessing.Process(target=f, args=(500000,))
            jobs.append(p)
            p.start()
            p.join()

        end = time.time()
        
        t = t + (end - start)
    t4=t/4

    t = 0

    for count in range(4):
        jobs = []
        start = time.time()
        for i in range(2):
            p = multiprocessing.Process(target=f, args=(1000000,))
            jobs.append(p)
            p.start()
            p.join()

        end = time.time()
        
        t = t + (end - start)
    t2=t/4

   
processes = [2,4,8,16,32]
times = [t2,t4,t8,t16,t32]
    
plt.figure(figsize=(15,15))
plt.plot(processes, times, linewidth=2.0, marker='o', color='k')
plt.grid(True)
plt.axis([2,32,0.8,1.8])
plt.title("Process level parallelism",fontsize=15)
plt.xlabel("Processes", fontsize=15)
plt.xscale('log',basex=2)
plt.ylabel("Time (seconds)", fontsize=15)
plt.legend()

plt.show()
    
```

# Notes

- Setup:
    - 1 datapoint is average of 4 runs
    - Total number of iterations of math.exp() loop are constant
- The performance sees slight degradation going from 2 to 4 processes, slightly sharper at 8 processes but drastic at 32.
